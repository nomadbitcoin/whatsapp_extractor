```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x*x
```

```python
# metodo mais simples e rapida pra plotar duas variaveis
plt.plot(x,y)
plt.plot(x, y, color='r') #alera a cor da linha

# plt.plot(x, y, 'r--') #alera a linha para tracejada e cor vermelho

plt.xlabel('eixo x ') #altera informacao do eixo x

plt.ylabel('eixo y ') #altera informacao do eixo y

plt.title('titulo') #altera o titulo 
```

```python
# plt.subplot(nrows=1, ncols=2) # os dois fazem a mesma coisa mas aqui esta explicito

#o 1 como ultimo argumento indica que estou fazendo alteracoes no primeiro grafico
plt.subplot(1,2, 1)
plt.plot(x, y,'r--')

#alteracoes sendo feitas no segundo grafico
plt.subplot(1,2, 2)
plt.plot(y, x,'g*-')
```

```python
# iremos trabalhar uma forma de se ter mais controle da plotagem atraves de um objeto
```

```python
#criamos uma figura vazia
fig = plt.figure()

#adicionamos um eixo 
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #possui 4 argumentos, 1 % de relacionamento com a esquerda da figura, 2 eh  % de baixo, 3 a % de largura, e por ultimo a altura
# esta a 10% do canto esquerdo, 10% da parte inferior e possui 80% de largura e 80% de altura 

axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3]) #posicionando ela dentro do outro grafico


axes1.plot(x, y)
axes1.set_xlabel('Eixo x')
axes1.set_title('Titulo')
axes2.plot(y, x)
```

```python
fig, ax = plt.subplots() #cria uma figura e cria um eixo

ax.plot(x, x**3, 'b--') #x ao cubo 
ax.plot(x, x**4, 'g-.') #x elevado a 4 potencia
ax.set_xlabel('x')
ax.set_label('y')
ax.set_title('Potencias de 3')
```

```python
fig, ax = plt.subplots(1,2)

#usando instancias de classe
ax[0].plot(x,y, 'b')
ax[0].set_title('Grafico 1')

ax[1].plot(x,y, 'b')
ax[1].set_title('Grafico 1')
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(10,3))

axes.plot(x,y, 'r--', label='x ^ 2')
axes.plot(y,x, 'b', label='x ^1/2')
axes.legend(loc=0) #loc define onde  vai estar posicionado a legenda
```

```python
fig, axes = plt.subplots(figsize=(10,10))

axes.plot(x**2, 'b', label='x ao quadrado', lw=10, alpha=0.5)
axes.plot(x, 'r', label='x na 4', linestyle='-.', lw=3)

#define os limites numericos para os eixos nas legendas embaixo do grafico
axes.set_xlim([0,len(x)]) 
axes.set_ylim([0,len(y)])

fig.legend()
```
