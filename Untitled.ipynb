```python
import os
# VERIFICA OS QUE JA ESTAO SALVOS NO DIRETORIO
path = './history/'
saved = []
for item in os.listdir(path):
    if item.endswith('.txt'):
        with open(path + item, 'r') as filee:
            data = filee.readlines()
            for line in data:
                if ('name') in line:
                    saved.append(line[line.find(':'):line.find('\n')])
```
